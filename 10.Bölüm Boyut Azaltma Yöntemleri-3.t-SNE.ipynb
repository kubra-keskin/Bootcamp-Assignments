```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
import time
```

```python
mnist = fetch_openml('Fashion-MNIST', version=1, cache=True)
```

```python
np.random.seed(123)

indices = np.random.choice(70000,10000)
X = mnist.data.iloc[indices].values / 255.0
y = mnist.target[indices].values
print(X.shape, y.shape)
y=[int(i) for i in y]
```

```python
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
plt.figure(figsize=(20,5))

for i in range(4):
    plt.subplot(1,4,i+1)
    plt.imshow(X_std[i].reshape(28,28), cmap="gray")
    plt.title("Resmin Etiketi: {}".format(y[i]))
plt.show()
```

```python
time_start = time.time()
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tsne_results = tsne.fit_transform(X)

print('t-SNE tamamlandı! Geçen zaman: {} saniye'.format(time.time()-time_start))
```

```python
plt.figure(figsize=(10,5))
plt.scatter(tsne_results[:, 0], tsne_results[:, 1],c=y)
plt.xticks([])
plt.yticks([])
plt.title("perplexity=40")
plt.axis('off')
plt.show()
```

```python
time_start = time.time()
perplexity_values = [10,50,100,150]

for i in range(0,4):
    tsne = TSNE(n_components=2, verbose=1, perplexity=perplexity_values[i], n_iter=300)
    tsne_results = tsne.fit_transform(X)

    print('t-SNE tamamlandı! Geçen zaman: {} saniye'.format(time.time()-time_start))

    plt.figure(figsize=(10,5))
    plt.scatter(tsne_results[:, 0], tsne_results[:, 1],c=y)
    plt.xticks([])
    plt.yticks([])
    plt.title("perplexity={}".format(perplexity_values[i]))
    plt.axis('off')
    plt.show()
```

perplexity değerleri açısından karşılaştırırsak kümeleri ayırmada en iyi işlemi gerçekleştiren perplexity=40 için olduğunu söyleyebiliri.

PCA ile kıyaslarsak:

```python
pca = PCA(n_components=2)
pca_components = pca.fit_transform(X_std)

plt.figure(figsize=(10,5))
plt.scatter(pca_components[:, 0], pca_components[:, 1],c=y)
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```

t-SNE yönteminin kümeleri ayırmada PCA'dan daha etkili olduğunu açıkça söyleyebiliriz.
